# Hierarchical Clustering

Let us use the iris dataset and create a dendrogram using all features

```python
iris_dset = load_iris ()
x = iris_dset['data']
print('the iris dataset has {} features'.format(x.shape[1]))
```

Now let us use the `dendogram` method in `scipy.cluster.hierarchy` to create a dendrogram

```python
plt.figure(figsize=(18, 7))
plt.title("Dendrograms")
res = sc.dendrogram(sc.linkage(x, method='ward'))
plt.title('Dendrogram')
plt.xlabel('Sample index')
plt.ylabel('Euclidean distance')
plt.show()
```
