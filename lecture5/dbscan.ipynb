```python
from sklearn.cluster import DBSCAN
from sklearn.datasets import make_blobs
import matplotlib.pyplot as plt
import numpy as np
```

# Evaluating the impact of Neighborhood radius size


## 1) First, let's generate and visualize some data...

```python
x, _ = make_blobs(n_samples=1000, centers=8, n_features=2, random_state=800)
plt.scatter(x[:,0],x[:,1])
plt.show()
```

## 2) Use scikit-learn to fit the DBSCAN lagorithm

Use a radius of 0.5, a minimum number of vectors in a neighborhood equal to 10, and use Euclidean distance

```python
dbscan = DBSCAN(eps=0.5, min_samples=10, metric='euclidean')
```

## 3) Fit the algorithm and plot the results

```python
labels = dbscan.fit_predict(x)
plt.scatter(x[:,0],x[:,1],c=labels)
plt.show()
```

```python
print('using a radius of {} and a minimum no of vectores {}, we\
 obtained {} clusters'.format(0.5,10,len(np.unique(labels))))
```

## 4) Cross-validate the radius
Use a grid-search appraoch


```python
radius = [0.2, 0.4, 0.6, 0.8]
fig, ax = plt.subplots(2,2)
c=0
for i in range(2):
  for j in range(2):
    dbscan = DBSCAN(eps=radius[c], min_samples=10, metric='euclidean')
    labels = dbscan.fit_predict(x)
    ax[i,j].scatter(x[:,0],x[:,1],c=labels)
    ax[i,j].set_title('radius {}'.format(radius[c]))
    c+=1
plt.tight_layout()
plt.show()
```
